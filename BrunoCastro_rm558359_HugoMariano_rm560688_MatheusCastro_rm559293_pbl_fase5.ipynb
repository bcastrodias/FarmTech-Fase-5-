**Análise estatística descritiva das variáveis, fornecendo as médias, os valores máximos e mínimos de cada variável.**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

features = ['Precipitation (mm day-1)', 'Specific Humidity at 2 Meters (g/kg)',
            'Relative Humidity at 2 Meters (%)', 'Temperature at 2 Meters (C)', 'Yield']

X = df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Método do Cotovelo para Determinar o Número de Clusters')
plt.xlabel('Número de Clusters')
plt.ylabel('Inércia')
plt.show()

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster_KMeans'] = kmeans.fit_predict(X_scaled)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)
df['PCA1'] = principal_components[:, 0]
df['PCA2'] = principal_components[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster_KMeans', data=df, palette='viridis')
plt.title('Clusters de K-means com PCA')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster_KMeans', y='Yield', data=df)
plt.title('Rendimento por Cluster (K-means)')
plt.xlabel('Cluster')
plt.ylabel('Rendimento (toneladas/ha)')
plt.show()
```

**Análise das distribuições dos clusteres**

O gráfico mostra a distribuição dos dados nos três clusters formados pelo algoritmo K-means, com a redução de dimensionalidade usando PCA para representar as variáveis em duas dimensões. O Cluster 0 (roxo) está concentrado em uma região próxima ao eixo vertical (PCA2), indicando características semelhantes entre essas observações. O Cluster 1 (verde-azulado) está mais isolado, sugerindo um comportamento distinto em comparação aos outros clusters. O Cluster 2 (amarelo) é mais disperso, o que indica maior diversidade nas observações desse grupo. Essa distribuição reflete como as variáveis influenciam a formação dos clusters, com alguns grupos apresentando características mais homogêneas e outros mais variadas.

**Modelos de Predição**

Regressão Linear

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

X = df[features]
y = df['Yield']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

cv_rmse = np.sqrt(-cv_scores.mean())



rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (Validação Cruzada): {cv_rmse}')
print(f'RMSE: {rmse}')
print(f'R²: {r2}')

```

Support Vector Machine

```python
from sklearn.svm import SVR

model = SVR()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

cv_rmse = np.sqrt(-cv_scores.mean())

rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (Validação Cruzada): {cv_rmse}')
print(f'RMSE: {rmse}')
print(f'R²: {r2}')

```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

cv_rmse = np.sqrt(-cv_scores.mean())

rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (Validação Cruzada): {cv_rmse}')
print(f'RMSE: {rmse}')
print(f'R²: {r2}')
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor


model = DecisionTreeRegressor(random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

cv_rmse = np.sqrt(-cv_scores.mean())

rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (Validação Cruzada): {cv_rmse}')
print(f'RMSE: {rmse}')
print(f'R²: {r2}')
```

XGBoost

```python
import xgboost as xgb


model = xgb.XGBRegressor(random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

cv_rmse = np.sqrt(-cv_scores.mean())

rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (Validação Cruzada): {cv_rmse}')
print(f'RMSE: {rmse}')
print(f'R²: {r2}')
```

**Comparação dos modelos**

O melhor modelo é o Random Forest, com RMSE (validação cruzada) de 6495.36 e R² de 0.9999, indicando excelente desempenho na previsão do rendimento, com um erro relativamente baixo e alta capacidade explicativa. Embora o modelo de Regressão Linear apresente um RMSE extremamente baixo em ambos os casos, o R² de 1.0 sugere sobreajuste, o que pode comprometer sua capacidade de generalizar para novos dados. O SVM teve o pior desempenho, com um R² negativo e um RMSE extremamente alto, indicando que não é adequado para este problema. Os modelos Decision Tree e XGBoost também apresentaram bom desempenho, mas o Random Forest mostrou o melhor equilíbrio entre baixo erro e alta explicação da variabilidade dos dados.

No entanto, os limites desse trabalho incluem a dependência de um número restrito de variáveis climáticas (como precipitação, umidade, temperatura), que podem não capturar toda a complexidade que afeta o rendimento das safras. Fatores como qualidade do solo, práticas agrícolas ou variáveis econômicas poderiam influenciar mais diretamente o rendimento, mas não foram considerados. Além disso, a análise não foi realizada em diferentes conjuntos de dados ou cenários, limitando a avaliação da generalização dos modelos. A ausência de uma análise temporal ou validação com dados de diferentes períodos também pode impactar a capacidade preditiva dos modelos a longo prazo.
