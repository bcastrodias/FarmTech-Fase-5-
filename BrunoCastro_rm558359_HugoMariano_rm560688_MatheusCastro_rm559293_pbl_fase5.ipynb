```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

features = ['Precipitation (mm day-1)', 'Specific Humidity at 2 Meters (g/kg)',
            'Relative Humidity at 2 Meters (%)', 'Temperature at 2 Meters (C)', 'Yield']

X = df[features]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('MÃ©todo do Cotovelo para Determinar o NÃºmero de Clusters')
plt.xlabel('NÃºmero de Clusters')
plt.ylabel('InÃ©rcia')
plt.show()

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster_KMeans'] = kmeans.fit_predict(X_scaled)

pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)
df['PCA1'] = principal_components[:, 0]
df['PCA2'] = principal_components[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster_KMeans', data=df, palette='viridis')
plt.title('Clusters de K-means com PCA')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster_KMeans', y='Yield', data=df)
plt.title('Rendimento por Cluster (K-means)')
plt.xlabel('Cluster')
plt.ylabel('Rendimento (toneladas/ha)')
plt.show()
```

**AnÃ¡lise das distribuiÃ§Ãµes dos clusteres**

O grÃ¡fico mostra a distribuiÃ§Ã£o dos dados nos trÃªs clusters formados pelo algoritmo K-means, com a reduÃ§Ã£o de dimensionalidade usando PCA para representar as variÃ¡veis em duas dimensÃµes. O Cluster 0 (roxo) estÃ¡ concentrado em uma regiÃ£o prÃ³xima ao eixo vertical (PCA2), indicando caracterÃ­sticas semelhantes entre essas observaÃ§Ãµes. O Cluster 1 (verde-azulado) estÃ¡ mais isolado, sugerindo um comportamento distinto em comparaÃ§Ã£o aos outros clusters. O Cluster 2 (amarelo) Ã© mais disperso, o que indica maior diversidade nas observaÃ§Ãµes desse grupo. Essa distribuiÃ§Ã£o reflete como as variÃ¡veis influenciam a formaÃ§Ã£o dos clusters, com alguns grupos apresentando caracterÃ­sticas mais homogÃªneas e outros mais variadas.

**Modelos de PrediÃ§Ã£o**

RegressÃ£o Linear

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

X = df[features]
y = df['Yield']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

cv_rmse = np.sqrt(-cv_scores.mean())



rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (ValidaÃ§Ã£o Cruzada): {cv_rmse}')
print(f'RMSE: {rmse}')
print(f'RÂ²: {r2}')

```

Support Vector Machine

```python
from sklearn.svm import SVR

model = SVR()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

cv_rmse = np.sqrt(-cv_scores.mean())

rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (ValidaÃ§Ã£o Cruzada): {cv_rmse}')
print(f'RMSE: {rmse}')
print(f'RÂ²: {r2}')

```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

cv_rmse = np.sqrt(-cv_scores.mean())

rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (ValidaÃ§Ã£o Cruzada): {cv_rmse}')
print(f'RMSE: {rmse}')
print(f'RÂ²: {r2}')
```

Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor


model = DecisionTreeRegressor(random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

cv_rmse = np.sqrt(-cv_scores.mean())

rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (ValidaÃ§Ã£o Cruzada): {cv_rmse}')
print(f'RMSE: {rmse}')
print(f'RÂ²: {r2}')
```

XGBoost

```python
import xgboost as xgb


model = xgb.XGBRegressor(random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cv_scores = cross_val_score(model, X_scaled, y, cv=5, scoring='neg_mean_squared_error')

cv_rmse = np.sqrt(-cv_scores.mean())

rmse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'RMSE (ValidaÃ§Ã£o Cruzada): {cv_rmse}')
print(f'RMSE: {rmse}')
print(f'RÂ²: {r2}')
```

Embora o modelo de RegressÃ£o Linear apresente um RMSE extremamente baixo em ambos os casos, o RÂ² de 1.0 sugere sobreajuste, o que pode comprometer sua capacidade de generalizar para novos dados. O SVM teve o pior desempenho, com um RÂ² negativo e um RMSE extremamente alto, indicando que nÃ£o Ã© adequado para este problema.
